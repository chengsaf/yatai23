```python
import matplotlib.pyplot as plt
import re
import os

# 正则表达式用于从文件名中提取 ID 号
pattern = re.compile(r'Fruit \((\d+)\)\.jpg')

# 存储识别为苹果的图片的 ID 号
apple_ids = []

# 遍历预测结果
for r in res:
    top_class_idx = r.probs.top1  # 获取 top1 类别索引
    top_class_name = r.names[top_class_idx]  # 获取类别名称

    if top_class_name == 'Apple':  # 检查 top 类别是否为“Apple”
        filename = os.path.basename(r.path)  # 从路径中获取文件名
        match = pattern.match(filename)
        if match:
            apple_ids.append(int(match.group(1)))  # 提取并存储 ID 号

# 绘制 ID 号的直方图
plt.hist(apple_ids, bins=20)  # bins 参数可根据需要调整
plt.title('Distribution of Apple IDs')
plt.xlabel('ID Number')
plt.ylabel('Frequency')
plt.show()
```
