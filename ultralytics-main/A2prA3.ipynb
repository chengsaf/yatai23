```python
import matplotlib.pyplot as plt
import re
import os

# æ­£åˆ™è¡¨è¾¾å¼ç”¨äºŽä»Žæ–‡ä»¶åä¸­æå– ID å·
pattern = re.compile(r'Fruit \((\d+)\)\.jpg')

# å­˜å‚¨è¯†åˆ«ä¸ºè‹¹æžœçš„å›¾ç‰‡çš„ ID å·
apple_ids = []

# éåŽ†é¢„æµ‹ç»“æžœ
for r in res:
    top_class_idx = r.probs.top1  # èŽ·å– top1 ç±»åˆ«ç´¢å¼•
    top_class_name = r.names[top_class_idx]  # èŽ·å–ç±»åˆ«åç§°

    if top_class_name == 'Apple':  # æ£€æŸ¥ top ç±»åˆ«æ˜¯å¦ä¸ºâ€œAppleâ€
        filename = os.path.basename(r.path)  # ä»Žè·¯å¾„ä¸­èŽ·å–æ–‡ä»¶å
        match = pattern.match(filename)
        if match:
            apple_ids.append(int(match.group(1)))  # æå–å¹¶å­˜å‚¨ ID å·

# ç»˜åˆ¶ ID å·çš„ç›´æ–¹å›¾
plt.hist(apple_ids, bins=20)  # bins å‚æ•°å¯æ ¹æ®éœ€è¦è°ƒæ•´
plt.title('Distribution of Apple IDs')
plt.xlabel('ID Number')
plt.ylabel('Frequency')
plt.show()
```
